```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
#Establish a connection to the police department's database
con = sqlite3.connect('/content/drive/MyDrive/sql-murder-mystery.db')
```

```python
#looking for data from crime scene report
query_1 = '''
SELECT *
FROM crime_scene_report
WHERE city = "SQL City"
ORDER BY date;
'''

pd.set_option('display.max_colwidth', None)
pd.read_sql_query(query_1, con)
```

```python
#there are 2 witnesses as per report
##checking personal details of both the witnesses
query_2 = '''
SELECT *
FROM person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number desc;
'''
pd.read_sql_query(query_2, con)[:1]
```

```python
query_3 = '''
SELECT *
FROM person
WHERE name like '%Annabel%' AND address_street_name = "Franklin Ave";
'''
pd.read_sql_query(query_3, con)
```

```python
#view the interview of both the witnesses taken after the murder
query_4 = '''
SELECT *
FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''
pd.read_sql_query(query_4, con)
```
